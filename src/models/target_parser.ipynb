```python
import numpy as np
import pandas as pd
```

### Parsing status codes to numbers

```python
def status_to_code(y):
    statusToNumb = {
        'Other'                 : 0,
        'Status_9_Substatus_8'  : 1,
        'Status_62_Substatus_7' : 2,
        'Status_62_Substatus_30': 3
    }
    y = np.array([statusToNumb[x] for x in y])
    #y['Status'] = y['Status'].map(statusToNumb)
    return y

def code_to_status(y):
    numbToStatus = {
        0: 'Other',
        1: 'Status_9_Substatus_8',
        2: 'Status_62_Substatus_7',
        3: 'Status_62_Substatus_30'
    }
    y = np.array([numbToStatus[x] for x in y])
    #y['Status'] = y['Status'].map(numbToStatus)
    return y

def status_to_description(y):
    statusToDescription = {
        'Other'                 : 'Other',
        'Status_9_Substatus_8'  : 'Generator heating - (Manual)',
        'Status_62_Substatus_7' : 'Feeding fault - (Diff. P-set/P-actual)',
        'Status_62_Substatus_30': 'Feeding fault - (Feeding safety circuit faulty)'
    }
    y = np.array([statusToDescription[x] for x in y])
    #y['Status'] = y['Status'].map(statusToDescription)
    return y

def description_to_status(y):
    descriptionToStatus = {
        'Other'                                          : 'Other',
        'Generator heating - (Manual)'                   : 'Status_9_Substatus_8',
        'Feeding fault - (Diff. P-set/P-actual)'         : 'Status_62_Substatus_7',
        'Feeding fault - (Feeding safety circuit faulty)': 'Status_62_Substatus_30'
    }
    y = np.array([descriptionToStatus[x] for x in y])
    #y['Status'] = y['Status'].map(descriptionToStatus)
    return y

def code_to_description(y):
    numbToDescription = {
        0: 'Other',
        1: 'Generator heating - (Manual)',
        2: 'Feeding fault - (Diff. P-set/P-actual)',
        3: 'Feeding fault - (Feeding safety circuit faulty)'
    }
    y = np.array([numbToDescription[x] for x in y])
    #y['Status'] = y['Status'].map(numbToDescription)
    return y

def description_to_code(y):
    descriptionToNumb = {
        'Other'                                          : 0,
        'Generator heating - (Manual)'                   : 1,
        'Feeding fault - (Diff. P-set/P-actual)'         : 2,
        'Feeding fault - (Feeding safety circuit faulty)': 3
    }
    y = np.array([descriptionToNumb[x] for x in y])
    #y['Status'] = y['Status'].map(descriptionToNumb)
    return y
```

```python
#y = pd.DataFrame([0,0,0,0,1,1,2,3,2,3,0,0,2], columns=['Status'])
y = [0,0,0,0,1,1,2,3,2,3,0,0,2]

print('Original y: \n', y, '\n\n')

y = code_to_status(y)
print('Code to status: \n', y, '\n\n')

y = status_to_code(y)
print('Status to code: \n', y, '\n\n')

y = code_to_description(y)
print('Code to description: \n', y, '\n\n')

y = description_to_code(y)
print('Description to code: \n', y, '\n\n')

y = code_to_status(y)
y = status_to_description(y)
print('Staus to description: \n', y, '\n\n')
```
